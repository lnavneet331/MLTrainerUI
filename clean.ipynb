```python
import pandas as pd

df = pd.read_csv('C:\\Users\\lnavn\Downloads\\archive (1)\\financial_regression.csv')

#drop the date column
df = df.drop('date', axis=1)
```

```python
df
```

```python
#drop values where everything is empty otherwise dont drop
df = df.dropna(how='all')
```

```python
# use simple imputer to fill missing values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)

```

```python
#split gold_close as another df
gold_close = df['gold close']
df = df.drop('gold close', axis=1)
```

```python
# save the data into different files
df.to_csv('x.csv', index=False)
gold_close.to_csv('y.csv', index=False)
```
